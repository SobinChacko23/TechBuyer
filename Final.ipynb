```python
import pandas as pd

# List of file paths
file_paths = [
    "C:\\SOBIN_CHACKO\\Servers_Storage.xlsx",
    "C:\\SOBIN_CHACKO\\BackupDrivers_Media.xlsx",
    "C:\\SOBIN_CHACKO\\Server_Memory.xlsx",
    "C:\\SOBIN_CHACKO\\Server_Processors.xlsx",
    "C:\\SOBIN_CHACKO\\Server_Power.xlsx",
    "C:\\SOBIN_CHACKO\\Laptops.xlsx",
    "C:\\SOBIN_CHACKO\\Desktop.xlsx",
    "C:\\SOBIN_CHACKO\\Components.xlsx",
    "C:\\SOBIN_CHACKO\\Switches.xlsx",
    "C:\\SOBIN_CHACKO\\Routers.xlsx",
    "C:\\SOBIN_CHACKO\\Accessories.xlsx",
    "C:\\SOBIN_CHACKO\\Servers.xlsx",
    "C:\\SOBIN_CHACKO\\Disk Storage and Enclosures.xlsx",
    "C:\\SOBIN_CHACKO\\MSA.xlsx",
    "C:\\SOBIN_CHACKO\\HP 3PAR.xlsx",
    "C:\\SOBIN_CHACKO\\Blade Modules and Switches.xlsx",
    "C:\\SOBIN_CHACKO\\Laptop_components.xlsx",
    "C:\\SOBIN_CHACKO\\Desktop Components.xlsx"
]

# List to store individual dataframes
dataframes = []

# Read each file into a dataframe and store it in the list
for file_path in file_paths:
    df = pd.read_excel(file_path)
    dataframes.append(df)

# Combine all dataframes into one
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df = combined_df.drop_duplicates()
```

```python

# Print the first few rows of the combined dataframe to verify
print(combined_df.head())
```

```python
# Clean the image_link column to remove any text after .jpg or .png
combined_df['image_link'] = combined_df['image_link'].str.replace(r'(\.jpg).*', r'\1', regex=True)
combined_df['image_link'] = combined_df['image_link'].str.replace(r'(\.png).*', r'\1', regex=True)
```

```python
# Optionally, save the combined dataframe to a new Excel file
combined_df.to_excel("C:\\SOBIN_CHACKO\\TechBuyer.xlsx", index=False)
```
